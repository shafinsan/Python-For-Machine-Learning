```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=sns.load_dataset('tips')
data
```

```python

sns.scatterplot(x='total_bill',y='tip',data=data[data['sex']=='Male'],marker='+',color='r',size='size',sizes=(20,200))
sns.scatterplot(x='total_bill',y='tip',data=data[data['sex']=='Female'],marker='+',color='g',size='size',sizes=(20,200))

plt.grid()
sns.despine()
plt.show()
```
