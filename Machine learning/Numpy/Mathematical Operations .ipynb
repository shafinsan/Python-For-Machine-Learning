```python
import numpy as np
a=[[1,2,3],[4,5,6],[7,8,9]]
b=[[11,12,13],[14,15,16],[17,18,19]]
a=np.array(a)
b=np.array(b)
print(a+b)
print(np.add(a,b))
print(".....................................................")
print(a-b)
print(np.subtract(a,b))
print(".....................................................")
print(a*b)
print(np.multiply(a,b))
print(".....................................................")
print(a/b)
print(np.divide(a,b))
print(".....................................................")
print(np.transpose(a))
```
