```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(y)

plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')

plt.plot(x,y)
plt.show()

```

```python
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('cos wave')
plt.plot(x,z)
plt.show()
```

```python
x=np.linspace(-10,10,200)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(-10,10,1000)
plt.plot(x,np.sin(x),'r+')
plt.plot(x,np.cos(x),'g+')
plt.show()
```

```python
x=np.linspace(-10,10,1000)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
people=['bangladesh','english','arabic','chinese','japanease']
number=[10,20,40,48,5]
ax.bar(people,number)
plt.xlabel('people')
plt.ylabel('number')
plt.title('graph')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
people=['bangladesh','english','arabic','chinese','japanease']
number=[10,20,40,48,5]
ax.pie(number,labels=people,autopct='%1.1f%%')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
x=10*np.linspace(0,1,10000)
y=x**2
#z=np.tan(x**2)
plt.xlabel('x-label')
plt.ylabel('y-label')
ax.scatter(y,x,c=x,cmap='Greens')
plt.show()
```
