```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('tips')
df1=df.corr(numeric_only=True)
```

```python
# We can use vmin,vmax,palette,cmap so on
# Heatmap works on only numarical values.... 
sns.heatmap(data=df1,annot=True,fmt='.1f')
plt.show()
```

```python
#joinplot only works on only numarical values also to find the relation between thos two numarical values relation

sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex',ratio=4,height=7, marginal_ticks=False)
plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex',ratio=4,height=7, marginal_ticks=True)
plt.show()
```

```python
#pairplot give us the grop all numarical values
sns.pairplot(df,kind='scatter',hue='sex',markers=['o','D'],height=5)
plt.show()
```

```python
sns.pairplot(df,kind='scatter',hue='sex',markers=['o','D'],height=5,corner=True)
plt.show()
```

```python
sns.pairplot(df,kind='scatter',hue='sex',markers=['o','D'],height=5,corner=True,diag_kind='kde')
plt.show()
```

```python
sns.pairplot(df,x_vars=['total_bill'],y_vars=['size'],hue='sex',kind='scatter')
plt.show()
```

```python
#displot using for only total count
sns.displot(x='tip',data=df,col='day',hue='sex',kde=True,rug=True,height=15,aspect=.4)
plt.show()
```
