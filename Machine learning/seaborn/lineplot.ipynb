```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=sns.load_dataset('tips')
data
```

```python
sns.lineplot(x='total_bill',y='tip',data=data,palette='Greens',hue='sex',
             marker='+',markeredgecolor='r',hue_order=['Female','male'])
plt.grid()
sns.despine()
plt.show()
```
