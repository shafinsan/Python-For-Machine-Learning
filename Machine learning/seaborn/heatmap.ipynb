```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=sns.load_dataset('tips')
data
data=data.drop(['sex','smoker','day','time'],axis=1).head(10)
data
sns.heatmap(data)
plt.show()
```

```python
newData=np.linspace(0,10,20).reshape(4,5)
sns.set_style('darkgrid')
sns.heatmap(newData,annot=True,annot_kws={'fontsize':15,'color':'b','fontfamily':'sans-serif'},
           xticklabels=False,robust=True,linecolor='b',linewidths=5)
plt.grid()
plt.show()
```

```python
data=data.head(100)
sns.set_style('darkgrid')
sns.heatmap(data,annot=True,annot_kws={'fontsize':15,'color':'b','fontfamily':'sans-serif'},
           xticklabels=False,robust=True,linecolor='b',linewidths=0)
plt.grid()
plt.show()
```
