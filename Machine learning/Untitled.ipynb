# Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

data1 = [1, 2, 3, 4, 5, 6, 7]
data2 = [2, 3, 4, 1, 5, 6, 7]

plt.plot(data1, data2)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import pandas as pd
myDataFrame=pd.DataFrame({'data1':data1,'data2':data2})
sns.lineplot(x='data1',y='data2',data=myDataFrame)
plt.show()
```

```python
load=sns.load_dataset('penguins')
#Make it more small bby
load=load.head(200)
load
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=load,
             hue='sex',style='sex',palette="Accent",
             markers=['o','>'],dashes=False,legend=False,size=50)
plt.grid()
plt.
plt.show()
```

```python
sns.barplot(x='flipper_length_mm',y='bill_length_mm',
            data=load,orient='v',errorbar=('ci',100),n_boot=1000
           ,saturation=10,alpha=0.5,
            capsize=0.8,width=0.8,errwidth=1,
            errcolor='red')
plt.show()
```

```python
load
sns.displot(data=load,x="flipper_length_mm",hue='sex',
            color='red',palette='Accent',alpha=0.5,kde=True,bins=[170,180,190,200,210,220,230])
plt.show()
```

```python

sns.scatterplot(x="flipper_length_mm",y='bill_depth_mm',
                data=load,hue='sex',
                palette='Accent',style='sex',
                markers={'Male':'*','Female':'o'},
                size='sex',sizes=(100,100)
               )
plt.show()
```
