```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python

df = sns.load_dataset('tips')
```

```python
df
```

```python
df1=df.corr(numeric_only=True)
```

```python
## lineplot it is typically used for continuous time-series
sns.lineplot(x='tip',y='total_bill',data=df,hue='sex',size=20,dashes=True,palette='Accent',markers=["+",">","o"])
sns.despine()
plt.grid()
plt.show()
```

```python
##barplot
data = {
    'Brand': ['LinkedIn', 'DHL', 'Google', 'FedEx', 'WhatsApp', 'Amazon', 'AliExpress', 'Apple', 'Maersk'],
    'Percentage': [52, 14, 7, 6, 4, 2, 1, 0.8, 0.8]
}
data=pd.DataFrame(data)
data

```

```python

sns.barplot(x='Percentage',y='Brand',data=data)
plt.show()
```

```python

sns.barplot(x='tip',y='total_bill',data=df,errorbar=('ci', 50), n_boot=2,errwidth=1,capsize=1)
plt.show()
```

```python
sns.countplot(x='sex',data=df,hue='day')
plt.show()
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,hue='day',color='w')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',inner='quart',fill=False,gap=.5,split=True,color='w')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=df,hue='sex')
plt.show()
```
