```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10,10,100)
plt.axes(projection='3d')
y=np.sin(x)
z=np.cos(x)

plt.xlabel('angle')
plt.ylabel('value')
plt.title('this is for practice')
plt.scatter(y,z,x,c=x,cmap='Blues')
plt.show()
```

```python
people=['bangladesh','english','arabic','chinese','japanease']
number=[10,20,40,48,5]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(people,number)
plt.show()
```

```python
people=['bangladesh','english','arabic','chinese','japanease']
number=[10,20,40,48,5]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.pie(number,labels=people,autopct='%1.1f%%')
plt.show()
```
